# Matplotlib (Draw figures in Python)

## Matplotlib: Python 2D plot library

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Scatter plot

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
area = np.pi * (15 * np.random.rand(n)) ** 2

plt.figure()
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

Line plot

```python
t = np.arange(0.0, 5.0, 0.2)

plt.figure()
plt.plot(t, t, "r--", t, t**2, "bs", t, t**3, "g^")
plt.show()
```

Line propreties
- linestyle or ls: ["-", "--", "-.", ":", "steps", "..."]
- linewidth or lw: any float values
- marker: ["+", ",", ".", "1", "2", "3", "4"]
- markersize or ms: float values
- markeredgecolor or mec: any matplotlib colors
- markeredgewidth or mew: float values in points
- markerfacecolor or mfc: any matplotlib color
- alpha: float

Bar chart

```python
people = ("Tom", "Dick", "Harray", "Slim", "Jin")
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.figure()
plt.bar(y_pos, performance, yerr=error, align="center", alpha=0.4)
plt.xticks(y_pos, people)
plt.ylabel("Performance")
plt.title("How fast do you want to go today?")
plt.show()
```

Pie chart

```python
labels = ["Frogs", "Hogs", "Dogs", "Logs"]
sizes = [15, 30, 45, 10]
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral"]
explode = (0, 0.1, 0, 0) # only explode the 2nd slice.

plt.figure()
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=90)
plt.show()
```

legend

```python
a = b = np.arange(0, 3, 0.02)
c = np.exp(a)
d = c[::-1]

plt.figure()
plt.plot(a, c, "k--", label="Model length")
plt.plot(a, d, "k:", label="Data length")
plt.plot(a, c + d, "k", label="Total message length")
legend = plt.legend(loc="upper center", shadow=True, fontsize="x-large")

legend.get_frame().set_facecolor("#00FFCC")
plt.show()
```

Text

```python
font = {
    "family": "serif", 
    "color": "darkred", 
    "weight": "normal", 
    "size": 16
}

x = np.linspace(0.0, 5.0, 100)
y = np.cos(2 * np.pi * x) * np.exp(-x)

plt.figure()
plt.plot(x, y, "k")
plt.title("Damped exponential decay", fontdict=font)
plt.text(2, 0.65, r"$\cos(2 \pi t) \exp(-t)$", fontdict=font)
plt.xlabel("time (s)", fontdict=font)
plt.ylabel("voltage (mV)", fontdict=font)

plt.subplots_adjust(left=0.15)
plt.show()
```
